```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
import joblib

def preprocess_input(data, label_encoders, scaler):
    df = pd.DataFrame(data)
    columns_to_drop = ['Insurance Provider', 'Admission Type', 'Billing Amount']
    df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
    for column in ['Gender', 'Blood Type', 'Medication']:
        df[column] = df[column].map(lambda s: label_encoders[column].transform([s])[0] if s in label_encoders[column].classes_ else -1)
    df = scaler.transform(df)
    return df

file_path = 'healthcare new (1).xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')
columns_to_drop = ["Date of Admission", "Doctor", "Room Number", "Discharge Date", "Hospital", "Test Results", "Name", "Insurance Provider", "Admission Type", "Billing Amount"]
df = df.drop(columns=columns_to_drop)

df[['Systolic', 'Diastolic']] = df['Blood Pressure (Systolic/Diastolic)'].str.split('/', expand=True)
df['Systolic'] = df['Systolic'].astype(int)
df['Diastolic'] = df['Diastolic'].astype(int)
df = df.drop(columns=['Blood Pressure (Systolic/Diastolic)'])
df = df.dropna()

print(df.head())

label_encoders = {}
for column in ['Gender', 'Blood Type', 'Medication']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Age Distribution')

plt.subplot(2, 2, 2)
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Gender Distribution')

plt.subplot(2, 2, 3)
sns.countplot(x='Blood Type', data=df, palette='plasma')
plt.title('Blood Type Distribution')

plt.subplot(2, 2, 4)
sns.countplot(x='Medical Condition', data=df, palette='inferno')
plt.title('Medical Condition Distribution')

plt.tight_layout()
plt.show()

X = df.drop(columns=['Medical Condition'])
y = df['Medical Condition']

le_target = LabelEncoder()
y = le_target.fit_transform(y)

scaler = StandardScaler()
X = scaler.fit_transform(X)

model = RandomForestClassifier(random_state=42)
model.fit(X, y)

joblib.dump(model, 'medical_condition_model.pkl')
joblib.dump(label_encoders, 'label_encoders.pkl')
joblib.dump(scaler, 'scaler.pkl')

def get_user_input():
    data = {
        "Age": [51],
        "Gender": ["Male"],
        "Blood Type": ["O-"],
        "Medication": ["Paracetamol"],
        "Oxygen Saturation (%)": [94.21725038082664],
        "Body Temperature (Â°C)": [36.0115054390603],
        "Blood Sugar (mg/dL)": [98.5227786652148],
        "Systolic": [104],
        "Diastolic": [81]
    }
    return data

model = joblib.load('medical_condition_model.pkl')
label_encoders = joblib.load('label_encoders.pkl')
scaler = joblib.load('scaler.pkl')

user_data = get_user_input()
preprocessed_data = preprocess_input(user_data, label_encoders, scaler)
prediction = model.predict(preprocessed_data)
predicted_condition = le_target.inverse_transform(prediction)

print(f"The predicted medical condition is: {predicted_condition[0]}")
```
